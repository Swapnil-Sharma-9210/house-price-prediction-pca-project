```python
numcols = df.select_dtypes(exclude='object')
```

```python
numcols.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
scaler.fit(numcols)
scaled_data=scaler.transform(numcols)
```

```python
scaled_data
```

```python
# generate the covariance matrix using 'cov' function
cov_mat = np.cov(scaled_data.T)

# as 'cov_mat' is a numpy array, select first five observations with [0:5]
print(cov_mat[0:5])
```

```python
eig_val, eig_vec = np.linalg.eig(cov_mat)
```

```python
eig_val
```

```python
eig_vec
```

```python
eig_val = list(eig_val)
eig_val.sort(reverse = True)
eig_val
```

```python
plt.plot(eig_val,'bx--')            
plt.plot(eig_val)      
plt.title('Screen Plot')
plt.show()   
```

```python
eigenvector = eig_vec[:,0:6]
eigenvector
```

```python
df_pca = pd.DataFrame(scaled_data.dot(eigenvector), columns= ['PC1','PC2','PC3','PC4','PC5','PC6'])
df_pca.head()
```

# Using SKLearn

```python
from sklearn.decomposition import PCA
```

```python
pca_price = PCA(n_components=6, random_state = 101)
pca_data = pca_price.fit_transform(scaled_data)
```

```python
PCA_df = pd.DataFrame(data = pca_data, columns = ['PC1', 'PC2','PC3','PC4','PC5','PC6'])
PCA_df.head()
```
